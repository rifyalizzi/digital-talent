```python
print("rata-rata berat beras adalah :", data.berat_beras.mean())
print("beras terberat tahun 2011:" , data.berat_beras.max())
print("beras teringan tahun 2011:", data.berat_beras.min())
```

```python
# data berdasarkan kecamatan
print("Berat beras tiap kecamatan")
data.groupby('kecamatan').mean()
```

```python
#  rata-rata beras tiap kelurahan 
data.groupby(['tahun', 'jumlah_kelurahan']).mean()
```

```python
jenis = data.groupby('kecamatan').mean()['berat_beras']
jenis = jenis.plot(kind='bar', title ="Rata-Rata Beras berdasarkan Kecamatan tahun 2011", legend=True, fontsize=12)
jenis.set_xlabel('kecamatan', fontsize=12)
jenis.set_ylabel('berat_beras', fontsize=12)
```
