```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## One to many relationship

#### 1. Report the account representative for each customer.

```python
run ("""
SELECT customername, contactfirstname, contactlastname, employeenumber, firstname, lastname
FROM customers INNER JOIN  employees ON salesrepemployeenumber = employeenumber;
""")
```

#### 2. Report total payments for Atelier graphique.

```python
run ("""
SELECT customers.customername, Sum(amount) AS Expr1
FROM customers INNER JOIN payments
ON customers.customernumber=payments.customernumber
WHERE customers.customername Like 'Atelier%'
GROUP BY customername;
""")
```

#### 3. Report the total payments by date

```python
run ("""
SELECT paymentdate, Sum(amount) AS total
FROM payments
GROUP BY paymentdate;
""")
```

#### 4. Report the products that have not been sold.

```python
run ("""
SELECT productcode, productname 
FROM products
WHERE productcode NOT IN (
SELECT DISTINCT productcode FROM orderdetails);
""")
```

#### 5. List the amount paid by each customer.

```python
run ("""
SELECT customername, Sum(amount) AS total
FROM payments INNER JOIN customers
ON payments.customernumber = customers.customernumber
GROUP BY customername;
""")
```

#### 6. How many orders have been placed by Herkku Gifts?

```python
run ("""
SELECT customername, COUNT(ordernumber) AS TRANSACTIONS
FROM orders INNER JOIN customers
ON orders.customernumber = customers.customernumber
WHERE customername LIKE "Herkku Gifts"
GROUP BY customername;
""")
```

#### 7. Who are the employees in Boston?

```python
run ("""
SELECT  lastname, firstname, city
FROM employees INNER JOIN offices
ON employees.officecode = offices.officecode
WHERE city LIKE "boston";
""")
```

#### 8. Report those payments greater than 100,000 Dollars Sort the report so the customer who made the highest payment appears first.

```python
run ("""
SELECT  customername, amount
FROM customers INNER JOIN payments
ON customers.customernumber = payments.customernumber
WHERE amount > 100000
ORDER BY amount desc;
""")
```

#### 9. List the value of 'On Hold' orders.

```python
run ("""
SELECT  SUM(quantityordered * priceeach) AS TOTAL
FROM orders INNER JOIN orderdetails
ON orders.ordernumber = orderdetails.ordernumber
WHERE status LIKE "On Hold";
""")
```

#### 10. Report the number of orders 'On Hold' for each customer.

```python
run ("""
SELECT customername, COUNT(ordernumber) AS TRANSACTIONS
FROM orders INNER JOIN customers
ON orders.customernumber = customers.customernumber
WHERE status LIKE "On Hold"
GROUP BY customername;
""")
```

## Many to many relationship

#### 1. List products sold by order date.

```python
run ("""
select p.productName, o.orderDate
from products p, orders o, orderdetails d
where d.productCode = p.productCode and o.orderNumber = d.orderNumber
order by o.orderDate asc;
""")
```

#### 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run ("""
select p.productName, o.orderDate
from products p, orders o, orderdetails d
where d.productCode = p.productCode and o.orderNumber = d.orderNumber and productName = '1940 Ford Pickup Truck'
order by o.orderDate desc;

""")
```

#### 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than \$25,000?

```python
run("""
    SELECT 
    customerName,
    orderNumber,
    amount
FROM
    orders o
        INNER JOIN
    customers c USING (customerNumber)
        INNER JOIN
    payments p USING (customerNumber)
WHERE
    amount > 25000
ORDER BY amount 
""")
```

#### 4. Are there any products that appear on all orders?

```python
run("""
    SELECT 
    productName,
    COUNT(orderNumber) AS JumlahOrder
FROM
    orderdetails od
        INNER JOIN
    products p USING(productCode)
        INNER JOIN
    orders o USING(orderNumber)  
GROUP BY productName      
ORDER BY JumlahOrder        
""")
```

#### 5. List the names of products sold at less than 80% of the MSRP.

```python
run ("""
select productName, buyPrice, MSRP
from products 
where buyPrice < (MSRP * .8);
""")
```

#### 6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)

```python
run ("""
select p.productName, p.buyPrice, d.priceEach
from products p, orderdetails d
where p.productCode = d.productCode and d.priceEach >= (p.buyprice*2);

""")
```

#### 7. List the products ordered on a Monday.

```python
run ("""
SELECT 
    orderDate,
    productCode,
    productName, 
    orderNumber
FROM 
    products p
        INNER JOIN
    orderdetails od USING (productCode)
        INNER JOIN
    orders o USING (orderNumber)
WHERE 
    DAYNAME(orderDate) = 'Monday'
ORDER BY orderDate
""")
```

#### 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run ("""
select distinct p.productName,d.quantityOrdered, p.quantityInStock, o.status
from products p, orderdetails d, orders o
where p.productCode = d.productCode and o.status = 'On Hold';
""")
```
